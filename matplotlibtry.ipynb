```python
#Import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   

import warnings
warnings.filterwarnings('ignore')
```

```python
#load datasets

data = pd.read_csv('precious1.csv')

#read data
data
```

```python
data.info()
```

```python
#checking duplicates of the data

data[data.duplicated()]
```

```python
#drop duplicates

data.drop_duplicates(inplace=True)

data.info()
```

```python
#checking for null values

data.isnull().sum()
```

```python
#fill in missing values in the Age column with mean Age 

data['Age'] = data['Age'].fillna(data['Age'].mean())

#fill in missing values in the Height column with median Height

data['Height'] = data['Height'].fillna(data['Height'].median())

#fill in missing values in the Weight column with median Weight 

data['Weight'] = data['Weight'].fillna(data['Weight'].median())

#fill in missing values in the Medal column with No Medal

data['Medal'] = data['Medal'].fillna('No Medal')

data.isnull().sum() 
```

```python
data.isnull()
```

checking the distribution of the data

```python
#distribution of the data

data.hist(figsize=(8,7))
plt.show()
```

```python
data.info()
```

```python
#converting datatypes

data['Sex'] = data['Sex'].astype('category')
data['Age'] = data['Age'].astype('int')
data['Team'] = data['Team'].astype('category')
data['NOC'] = data['NOC'].astype('category')
data['Medal'] = data['Medal'].astype('category')
```

EXPLORATORY DATA ANALYSIS

```python
#distribution of the Age column
plt.figure(figsize=(18, 6))
plt.subplot(1,3,1)
data['Age'].hist(bins=20, color='orange', edgecolor='black', figsize=(12, 6))
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')


#distribution of the Weight column
plt.subplot(1,3,2)
data['Weight'].hist(bins=20, color='brown', edgecolor='black')
plt.title('Distribution of Weight')
plt.xlabel('Weight')
plt.ylabel('Frequency')


#distribution of the Height column
plt.subplot(1,3,3)
data['Height'].hist(bins=20, color='grey', edgecolor='black')
plt.title('Distribution of Height')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

FEATURE ENGINEERING

```python
#calculate the Body Mass Index of the athlete

data['BMI'] = data['Weight']/(data['Height'])**2
data
```

```python
#Relationship between Height and Body Mass Index

plt.figure(figsize=(18, 6))
plt.subplot(1,3,1)
plt.scatter(data['Height'], data['BMI'], color='black')
plt.title('Height vs BMI')
plt.xlabel('Height')

#Relationship between Sex and Body Mass Index

plt.subplot(1,3,2)
plt.scatter(data['Sex'], data['BMI'], color='green')
plt.title('Sex vs BMI')
plt.xlabel('Sex')

#Relationship between Age and Body Mass Index

plt.subplot(1,3,3)
plt.scatter(data['Age'], data['BMI'], color='yellow')
plt.title('Age vs BMI')
plt.xlabel('Age')

plt.show()
```

```python
#relationship between Weight and Height

plt.figure(figsize=(18, 6))
plt.subplot(1,3,1)
plt.scatter(data['Weight'], data['Height'], color='orange')
plt.title('Weight vs Height')
plt.xlabel('Weight')

#Relationship between Age and Weight

plt.subplot(1,3,2)
plt.scatter(data['Age'], data['Weight'], color ='green')
plt.title('Age vs Weight')
plt.xlabel('Age')

#Relationship between Sex and Weight

plt.subplot(1,3,3)
plt.scatter(data['Sex'], data['Weight'], color='black')
plt.title('Sex vs Weight')
plt.xlabel('Sex')
```

```python
#Calculate the correlation coefficients between numerical variables (Age, Height, Weight)

correlation_matrix = data[['Age', 'Height', 'Weight', 'BMI']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')

plt.show()
```

```python
#Plot line charts to visualize how certain variables Weight have changed over different years

# Grouping the data by 'Year' and calculating the mean weight for each year
yearly_weight_mean = data.groupby('Year')['Weight'].mean()

# Plotting the line chart
plt.figure(figsize=(10, 6))
yearly_weight_mean.plot(kind='line', marker='o', color='orange', linestyle='-')
plt.title('Average Weight Over Different Years')
plt.xlabel('Year')
plt.ylabel('Average Weight')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Grouping the data by 'Year' and calculating the mean height for each year
yearly_height_mean = data.groupby('Year')['Height'].mean()

# Plotting the line chart
plt.figure(figsize=(10, 6))
yearly_height_mean.plot(kind='line', marker='o', color='green', linestyle='dotted')
plt.title('Average Height Over Different Years')
plt.xlabel('Year')
plt.ylabel('Average Height')
plt.grid(True)
```
